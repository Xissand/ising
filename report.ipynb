Типичный вид решетки на малых температурах - все спины ориентированы в одном направлении.

```python
main.coolplot(50,0.1)
```

При этом можно наблюдать метастабильные состояния - спины ориентированы в одном направлении, но полоса (горизонтальная или вертикальная) ориентированна в другом направлении.

```python
main.coolplot(50,0.1)
```

Типичный вид решетки на больших температурах - спины перемешаны.

```python
main.coolplot(50, 4)
```
